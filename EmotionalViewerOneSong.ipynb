#### **Paso 2: Proceso de cargue de percepción emocional dinámica por diferentes anotadores en MediaEval.**

```python
import pandas as pd
import matplotlib.pyplot as plt

db_name= 'http://104.237.5.250/datasets/DEAM_Annotations/annotations/annotations%20per%20each%20rater/dynamic%20(per%20second%20annotations)/valence/'+str(song)+'.csv'

# Leemos y obtenemos el dataframe (percepciones emocionales)
df= pd.read_csv(db_name)
```

#### **Paso 3: Proceso de visualización de percepción emocional dinámica en valencia para un solo anotador**

```python
# Visualización de un proceso de evaluación de valence para una canción.
plt.figure()
plt.title("Variación temporal de la dimensión emocional")
plt.xlabel("Time window")
plt.ylabel('Valence')
print("Total de evaluaciones:",len(df.values))
print("Vector valencia del primer anotador:\n")
print(df.values[0])
plt.plot(df.values[0,1:]) # omitimos la primera posición (id anotador)
```
