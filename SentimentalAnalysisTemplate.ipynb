**Paper:** XLM-T: A Multilingual Language Model Toolkit for Twitter.

**Git Repo:** XLM-T official repository.

**Paso 1: Importar Librerías**

```python
from transformers import pipeline
model_path = "cardiffnlp/twitter-xlm-roberta-base-sentiment"
sentiment_task = pipeline("sentiment-analysis", model=model_path, tokenizer=model_path)
```

**Paso 2: Inferir sentimiento**

```python
# Comentario de ejemplo
print(sentiment_task("Arley Agudelo - La sensacion de la cancion la noto revolucionaria o futurista con un ritmo interesante. La letra trata de realizar un llamado de conciencia o un punto de vista nuevo. Lo que no me gusto es que es larga la cancion."))
```
